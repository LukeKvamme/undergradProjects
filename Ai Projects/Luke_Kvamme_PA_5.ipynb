**Question 1**. 

Given a body of text  with $n$  <u>_unique_</u>  tokens (say The Comedy of Errors by Shakespeare with ~3707 unique tokens) create an $n \times n$ matrix $Cooccurences$ such that $Cooccurences[i, j]$ is equal to the number of times the word $i$ <u>_preceeds_</u> word $j$ within a window length of $k$ tokens. 

For example, `text = "To be or not to be"` has $n=4$; for $k = 2,~Cooccurence_{~text}~$should be:

<br/>

\begin{bmatrix} 
	 & to & be & or & not \\
	to & 0 & 2 & 1 & 0\\
	be & 0 & 0 & 1 & 1\\
	or & 1 & 0 & 0 & 1\\
	not & 1 & 1 & 0 & 0\\
	\end{bmatrix}

```python
import requests

def tokenize(text):
  import spacy
  nlp = spacy.load("en_core_web_sm")
  all_tokens = [token.text.lower() for token in nlp(text)]
  vocab = set(all_tokens)
  return all_tokens, vocab

response = requests.get('https://shakespeare.folger.edu/downloads/txt/the-comedy-of-errors_TXT_FolgerShakespeare.txt')

text = response.content.decode()

all_tokens, vocab = tokenize(text)
```

```python
len(all_tokens), len(vocab)
vocabList = list(vocab)
```

```python
import pandas as pd

df = pd.DataFrame(index = vocab, columns = vocab)
df.head()
```

```python
df[:] = 0

def kthMatrix(k):
  kthList = [-1] * k

  for word in all_tokens:
    for prevWord in kthList:
      if prevWord == -1:
        continue
      df[word][prevWord] = df[word][prevWord] +1

    kthList.pop(0)
    kthList.append(word)

kthMatrix(2)
df.head()
```
